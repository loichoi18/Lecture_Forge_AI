## Inspect transcript stats

```python
import re

words      = transcript.split()
sentences  = re.split(r'[.!?]+', transcript)
paragraphs = [p for p in transcript.split('\n\n') if p.strip()]

print(f"Word count       : {len(words):,}")
print(f"Sentence count   : {len(sentences):,}")
print(f"Paragraph count  : {len(paragraphs):,}")
print(f"Avg words/sent   : {len(words)/max(len(sentences),1):.1f}")

# Estimate reading time
read_minutes = len(words) / 200
print(f"\nEstimated reading time: {read_minutes:.1f} minutes")
print(f"Estimated lecture length: ~{read_minutes * 2:.0f} minutes (speaking pace)")
```
